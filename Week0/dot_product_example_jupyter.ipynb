# Dot Product Example (Skalar produkt)

The dot product of two vectors is defined as:

$$\mathbf{a}\cdot\mathbf{b} = a_x b_x + a_y b_y$$

It can also be written as:

$$\mathbf{a}\cdot\mathbf{b} = |\mathbf{a}|\,|\mathbf{b}|\cos\theta$$

```python
print("Hello world")
```

```python
import numpy as np

# Define two vectors
a = np.array([2, 3])
b = np.array([4, -1])

# Compute dot product
dot = np.dot(a, b)

# Compute lengths
len_a = np.linalg.norm(a)
len_b = np.linalg.norm(b)

# Compute angle (degrees)
theta_deg = np.degrees(np.arccos(dot / (len_a * len_b)))

# Print nicely
print(f"Dot product: {dot}")
print(f"Length of a: {len_a:.3f}")
print(f"Length of b: {len_b:.3f}")
print(f"Angle between vectors: {theta_deg:.2f}°")
```
